# Problem Statement
<font size="4">How might we predict a country’s **protein supply per capita** through the analysis of social, environmental, and economic factors such as **population, gross domestic product (GDP), temperature change, food consumer price index (CPI), and import, export and production quantity**?</font>

<a id='problem'></a>
## Overview About the Problem

An individual has food security if he/she has physical, social and economic access to sufficient, safe and nutritious food that meets their personal preferences and dietary requirements for an active and healthy life. 

To measure this, the team decided on protein supply as a good measure of food security as it is the best food related indicator that is greatly subject to a person/country’s wealth. A study by the National Library of Medicine in 2020$^1$ revealed that protein intake per weight in upper income households are the highest among all the income groups, thus suggesting that it is the **nutrient most subjected to disparity due to price**. 

As such, the team decided to study the relationship between countries’ protein supply at the consumption stage of the supply chain, against their environmental, economic and social climate. 

<a id='persona'></a>
## User Persona: Humanitarian Organization
The team decided to craft a user persona centred around a **Humanitarian Organization or UN Organization** that provides nutritional aid to poorer countries or regions experiencing difficulties in protein production/acquisition.   

The model is useful to predict the **average protein supplied to each person** in a country accounting for the **differences in their socioeconomic conditions**, in terms of the predictors the team have chosen. This allows it to be applicable across different theatres. The protein supply can then be used to **predict an appropriate amount of aid provided by our user persona** to supplement the protein intake to recommended levels. 

## Assumptions

The team narrowed down certain predictors’ data to reduce redundancy and streamline our analysis. 
In the analysis, protein refers only to meat, and not to other types of food such as live animals, eggs and dairy which also contains protein. This assumption was applied to the import, export and production quantities of each country.

# Building Model


## Chosen Model: Multiple Linear Regression

Our chosen model is that of multiple linear regression (MLR). Multiple Linear Regression is used as the hypothesis of the protein supplied is not merely a straight line. This is due to complex nature of plotting protein supply which is a combination of a plethora of variables (predictors) from differing social, environmental, and economic factors in each country.    

In this case, our predictors are our various $X_i$ independent variables and the protein supplied per capita is our dependent variable. 

Each of these predictors have their own unknown weight which they contribute to the protein supply, and some that were previously hypothesized to have an effect of protein supply by the team may be completely independent or have negligible effects on our model. These predictors will be evaluated in by the team's chosen metric, discussed below, and then removed to improve the MLR model.

## Mathematics of MLR
In our model, we hypothesize that the protein supplied in g per capita to people in a country is can be predicted by the multiple linear regression function $\hat{y}$ as follows:

$$\hat{y}(x) =  \hat{\beta}_0 + \hat{\beta}_1 x_1 + \hat{\beta}_2 x_2 + \ldots + \hat{\beta}_n x_n$$  

#### <center> OR </center>  

$$\mathbf{\hat{y}} = \mathbf{X} \times \mathbf{\hat{b}}$$

<center> where $\mathbf{X}$ is a matrix representing the predictors $x_0, x_1, x_2, ..., x_n$ for each data points $x^i$</center>
<center> and $\hat{b}$ is the column vector representing the multiple $\hat{\beta_i}$ that best fits the training data </center>
  
  
The cost function of our model is the following, which is derived from the sum of squared errors, i.e $\frac{1}{2}$MSE:
$$J(\hat{\beta}_0, \hat{\beta}_1) = \frac{1}{2m}\Sigma^m_{i=1}\left(\hat{y}(x^i)-y^i\right)^2$$

Using calculus to minimize our cost J against $\hat{b}$, we can derive our gradient descent update equation as the following:
$$ \mathbf{\hat{b}} = \mathbf{\hat{b}} - \alpha\frac{1}{m} \mathbf{X}^T \times (\mathbf{X}\times \mathbf{\hat{b}} - \mathbf{y}) $$

With this, we get a column vector, $\hat{b}$, of various $\hat{\beta_i}$ for each predictor $x_i$ which best fits our data.  
  
To check if we have sufficiently minimized our cost throughout the machine learning, we can plot our cost at iteration, J, against iteration number.

## Hyperparameters for training
The model is trained on 70% of the dataset, and tested on the remaining 30%.  

$\hat{b}$ is initalized as a column vector of all zeros, and represents the starting guess for various $\hat{\beta_i}$.

Iterations is set to 1500, which from the cost graph sufficiently minimizes the cost throughout gradient descent.

```python
def calc_linear(X, beta):
    return np.matmul(X, beta)

def compute_cost(X, y, beta):
    J = 0
    m = X.shape[0]
    error = calc_linear(X, beta) - y
    error_sq = np.matmul(error.T, error)
    J = (1/(2*m))*error_sq
    J = J[0][0]
    return J

def gradient_descent(X, y, beta, alpha, num_iters):
    m = X.shape[0] 
    J_storage = np.zeros((num_iters, 1))
    
    for i in range(num_iters):
        beta = beta - (alpha / m) * np.matmul(X.T, (calc_linear(X, beta) - y))
        J_storage[i] = compute_cost(X, y, beta)
    return beta, J_storage
```

# Evaluating the Model

Metrics wise there are 3 that works in linear regression (excluding their upgraded variations that work in multiple linear regression):
- R squared (Linear Regression only) —> Adjusted R squared (Both Single and Multiple Linear Regression)
- Mean Squared Error (MSE) —> Root Mean Squared Error (RMSE) (i.e just corrected for the squaring of error)
- Mean Absolute Error (MAE)

MSE, RMSE and MAE measure the error of the model against the actual data. All 3 metrics range from 0 to $\infty$, and is negatively-oriented (i.e the lower it is the better).
  
## R-Squared
R squared measures the fit of the model by explaining how much of a variation of a dependent variable is explainable by independent variable in regression model. R squared only works on single linear regression, as it increases or remains the same, making the model seem like a better fit than it actually is.  

R-Squared is positive-oriented (i.e higher is better fit), and ranges from 0 to 1.  

Instead, we use **Adjusted R-Squared** which adjusts for degree of freedom of data, and is unaffected by irrelevant predictors.  
For p predictors and n sample size: 
$$ R^2_{\mathrm{adj}} = 1 - \frac{\mathrm{RSS}/(n-p)}{\mathrm{TSS}/(n-1)} $$
#### <center> OR </center>  
$$ R^2_{\mathrm{adj}} = 1-\frac{(1-R^2)(n-1)}{(n-p-1)} $$

Adjusted R-Squared can take negative values, which hints to a bad fit. This means that bad predictors will have to be removed.

## Mean Squared Error
MSE and RMSE are the same in distribution, but RMSE is scaled down in value. 

Mean Squared Error = Mean of squared error = $ \frac{1}{n}\sum_{i=1}^{n}(\hat{y}(x^i)-y_i)^2 $  
Root Mean Squared Error = $ \sqrt{MSE} $
  
MSE and RMSE are sensitive to outliers. This is as it gives higher weight to larger errors (residuals), and is better used when large errors are particularly undesirable. 
* Since we desire to minimize RMSE, we want to over represent large errors.

This is sensitivity is due to the squaring of the errors, which we can see mathematically from how:
$$ (\frac{b}{a})^2 > \frac{b}{a} \text{, where b is the large error and a is the small error}$$


## Mean Absolute Error
MAE = Mean of absolute error = $\frac{1}{n}\sum_{i=1}^{n}|\hat{y}(x^i)-y_i|$
MAE treats every error equally and just finds their average.  

However, absolute values are undesirable in mathematics as absolute value is a piecewise function which are subject to both algebra and a geometric interpretation of piecewise graph.

$$ \lvert x \rvert = \left\{
        \begin{array}{ll}
            -x & \quad x < 0 \\
            x & \quad x \geq 0
        \end{array}
    \right. $$

## Main Metric of Choice: RMSE
Given our model of protein supply is worse off having larger errors due to outliers in the data, as well as the ease of mathematical calculations, we will be choosing RMSE as our main metric.

Adjusted R-Squared will be used to further evaluate the fit of our model.


#### References: 
* [Medium: RMSE vs MAE - Which metric is better](https://medium.com/human-in-a-machine-world/mae-and-rmse-which-metric-is-better-e60ac3bde13d#:~:text=RMSE%20has%20the%20benefit%20of,then%20MAE%20is%20more%20appropriate.)
* [Medium: Comparing robustness of MAE, MSE and RMSE](https://towardsdatascience.com/comparing-robustness-of-mae-mse-and-rmse-6d69da870828)
* [Derivation of R-squared and adjusted R-squared](https://statproofbook.github.io/P/rsq-der.html)

```python
def r2_score(y, ypred):
    y_mean = y.mean()
    ss_res = np.sum((y - ypred)**2)
    ss_tot = np.sum((y - y_mean)**2)
    return 1 - ss_res/ss_tot

def mean_squared_error(target, pred):
    n = target.shape[0]
    rss = np.sum((target - pred)**2)
    mse = rss / n
    return mse

def mean_absolute_error(target, pred):
    n = target.shape[0]
    sum_abs = np.sum(np.abs(target - pred))
    return sum_abs / n

def root_mean_squared_error(target, pred):
    return np.sqrt(mean_squared_error(target, pred))

def adj_r2_score(y, ypred, predictor_no):
    # cannot determine predictor_no from y alone, need to passed in
    n = y.shape[0]  # sample size, i.e rows of result
    r2 = r2_score(y, ypred)
    return 1 - ((1-r2)*(n-1))/(n-predictor_no-1)

# Evaluation of model using RMSE and adjusted R2 score
def get_metrics(target, pred, predictor_count):
    model_rmse_score = root_mean_squared_error(target, pred)
    model_adj_r2_score = adj_r2_score(target, pred, predictor_count)
    model_r2_score = r2_score(target, pred)
    model_mae_score = mean_absolute_error(target, pred)
    
    return pd.DataFrame({"RMSE": [model_rmse_score],
                         "MAE": [model_mae_score],
                         "R-Squared": [model_r2_score],
                         "Adjusted R-Squared": [model_adj_r2_score]
                        })
```

```python
# put Python code to evaluate the model and to visualize its accuracy
df_metric = get_metrics(target, pred, predictor_count)
display(df_metric)
```

# Improving the Model

## Improvement 1: Removal of outliers
From our data, we can see certain countries protein supply are potentially far from the rest.  

For example, Zimbabwe has the lowest protein supply per g per pax of all other countries, of 36g, as a result of rapid inflation shown by their high CPI of >6.  
On the other extremity, Iceland has the highest protein supply per g per pax of 135g.
  
We can arbitarily define outliers as protein supply values less than (Q1 - 1.5 IQR) or protein supply values more than (Q3 + 1.5 IQR). The functions to find these are defined below.
  
This has two implications that allows us to get a better model:  
1. Our chosen cost metric, RMSE, is more sensitive to extreme values. By removing outliers, we can lower our RMSE.
2. Our cost function J is equal to $\frac{1}{2}$ MSE which like MSE is sensitive to outliers that can amplify the large error or underrepresent small errors due to the squaring of the errors.

### Double Check: Box and Whiskers Diagram
A boxplot will clearly show the five-number summary, that is minimum, quartile 1, mean, quartile 3 and maximum of the data. Any outliers can be detected from this graph.

### Conclusion of outliers
From the code and box plot below, there are apparently **no outliers** and all countries in our cleaned dataset falls **within the lower and upper ranges of 8.18 g and 165.3 g of protein** supplied per pax. 

## Improvement 2: Removal of irrelevant / low correlation predictors  

### Methodology: Brute Force and Check Metrics
For all 7 of the predictors chosen by the team, generate every combination of predictors to be considered in the MLR model.  
The MLR model is then trained and tested on the entire dataset (i.e no train-test split), to keep the metrics consistent. 
Then calculate the metrics for each and every combination, and get the lowest RMSE.  

This allows the detection of the best set of predictors that lead to the most accurate model.

### Conclusion of irrelevant predictors
Based on the table of RMSE metrics, we will be removing **population (per 1000)** as our predictor(s). This is as the lowest RMSE of 9.91 is obtained without it out of all the predictor combinations.

```python
# Improvement 1: Removal of outliers
def removeoutliers(df, target_col):
    q3 = np.quantile(df[target_col], 0.75)
    q1 = np.quantile(df[target_col], 0.25)
    iqr = q3 - q1
 
    lower_range = q1 - 1.5 * iqr
    upper_range = q3 + 1.5 * iqr
    
    print(f"{target_col}\nLower range: {lower_range} \nUpper range: {upper_range}")
    
    outlier_mask = (df[target_col] < lower_range) & (df[target_col] > upper_range)
    
    filtered_df = df.loc[~outlier_mask]
    dropped_df = df.loc[outlier_mask]
    return filtered_df, dropped_df

df_rem_outliers, df_outliers = removeoutliers(df_normalized, "Protein Supply per g per pax (y)")
display(df_rem_outliers)
display(df_outliers)
```

```python
# To double confirm outliers, box plot of Protein Supply per g per pax should have no outliers
sns.boxplot(data=df_normalized, x="Protein Supply per g per pax (y)")
```

```python
# Improvement 2: Removal of irrelevant / low correlation predictors

from itertools import combinations, chain

def powerset(iterable):
    "powerset([1,2,3]) --> () (1,) (2,) (3,) (1,2) (1,3) (2,3) (1,2,3)"
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(len(s)+1))


all_feature_names = ["Population (per 1000)", "Temperature Change (degrees celsius)", "Consumer Price Index", "GDP per capita (USD millions)", "Import Quantity (tonnes per 1000 people)", "Export Quantity (tonnes per 1000 people)", "Production Quantity(tonnes per 1000 people)"]
df_all_metrics = pd.DataFrame(columns=[*all_feature_names, "RMSE", "MAE", "R-Squared", "Adjusted R-Squared"])
all_feature_combo = powerset(all_feature_names)

# Brute force through every combination of predictors
for feature_names in all_feature_combo:
    feature_names = list(feature_names)
    target_names = ["Protein Supply per g per pax (y)"]
    df_features_norm, df_target = get_features_targets(df_normalized, feature_names, target_names)
    predictors = df_features_norm.columns
    predictor_count = len(predictors)
    X = prepare_feature(df_features_norm)
    target = prepare_target(df_target)
    iterations = 2000
    alpha = 0.01
    beta = np.zeros((predictor_count + 1,1))
    
    beta, J_storage = gradient_descent(X, target, beta, alpha, iterations)
    target = prepare_target(df_target)  # Actual y_i values
    pred = predict(df_features_norm, beta)  # Predicted y_hat value

    df_metric = get_metrics(target, pred, predictor_count)
    df_features_bool = pd.DataFrame({feature:[1] for feature in feature_names})
    new_df_row = df_features_bool.join(df_metric, how="left")
    df_all_metrics = pd.concat([new_df_row, df_all_metrics], sort=False, ignore_index=True)

df_all_metrics.fillna(0, inplace=True)
```

```python
# Find the lowest RMSE and max adjusted r2
display(df_all_metrics)

mask1 = df_all_metrics["Population (per 1000)"] == 1
mask2 = df_all_metrics["Temperature Change (degrees celsius)"] == 1
mask3 = df_all_metrics["Consumer Price Index"] == 1
mask4 = df_all_metrics["GDP per capita (USD millions)"] == 1
mask5 = df_all_metrics["Import Quantity (tonnes per 1000 people)"] == 1
mask6 = df_all_metrics["Export Quantity (tonnes per 1000 people)"] == 1
mask7 = df_all_metrics["Production Quantity(tonnes per 1000 people)"] == 1

min_rmse_predictors = df_all_metrics.iloc[df_all_metrics["RMSE"].idxmin(), :]
max_adj_r2_predictors = df_all_metrics.iloc[df_all_metrics["Adjusted R-Squared"].idxmax(), :]


display(min_rmse_predictors)
display(max_adj_r2_predictors)
display(df_all_metrics.tail(7))
```
